**Task1: Run Binary logit**

To do this task, you have to create another dataset with only specified columns that will be used for first model

```python
df_train = df[['MBR_binary','SBR_binary', 'paid_price','purchases', 'brand_switch', 'feature', 
                     'display', 'basket_size', 'residence_status', 'HH_Income', 'NumberMembers', 'MH_Working_Hours', 'MH_Edu', 'FH_Working_Hours', 'FH_Edu']]
```

```python
# dmatrices function takes formula in format y = x1+ x2+ X3 ...
# Therefore we are extracting column names and joining with '+'
features = " + ".join(df_train.drop(columns=["SBR_binary","MBR_binary"]).columns)
```

DV = Memory-Based Reference Price Shoppers

```python

# This line will return X and y values as dataframes to fit into a linear model.
y, X = dmatrices('MBR_binary ~' + features, df_train, return_type='dataframe')

# Run Linear regression model
mod = sm.Logit( y, X)

# fit the data into model
res = mod.fit()

# print the summary
print(res.summary())
```

DV = Stimulus-Based Reference Price Shoppers  

```python
# This line will return X and y values as dataframes to fit into a linear model.
y, X = dmatrices('SBR_binary ~' + features, df_train, return_type='dataframe')

# Run Linear regression model
mod = sm.Logit( y, X)

# fit the data into model
res = mod.fit()

# print the summary
print(res.summary())
```
